# Various Audio Effects

The goal of this project is to first implement various audio effects in hopes of implementing them on a realtime DSP unit for human voice modulation.

## A test sample

Below is a test audio sample in the time domain

```python
sampleFreq,sampleData = wavfile.read("test.wav")
N=sampleData.shape[0]
sampleDuration=N/sampleFreq
print(N)
t = np.arange(start=0,stop=sampleDuration,step=sampleFreq**(-1))
plt.plot(t,sampleData)
```

In the frequency domain:

```python
sampleDataSpectral = fft.fft(sampleData[:,0])
sampleFreqSpace = fft.fftfreq(N,sampleFreq**(-1))
plt.plot(sampleFreqSpace,abs(sampleDataSpectral))
plt.xlims=([-20000,20000])
```
